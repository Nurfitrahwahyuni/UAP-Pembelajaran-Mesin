```python
# ==========================================
# 1. SETUP & IMPORT LIBRARIES
# ==========================================
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import random

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Conv2D, MaxPooling2D, Flatten, Dense,
                                      Dropout, BatchNormalization, GlobalAveragePooling2D)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.applications import EfficientNetB7, MobileNetV2
from tensorflow.keras.applications.efficientnet import preprocess_input as preprocess_b7
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as preprocess_mnet

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from google.colab import files

# ==========================================
# 2. SETUP KAGGLE API & DOWNLOAD DATASET
# ==========================================
!mkdir -p ~/.kaggle
!cp /content/kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

!kaggle datasets download -d trolukovich/food11-image-dataset
!unzip food11-image-dataset.zip -d /content/dataset

# ==========================================
# 3. DATASET EXPLORATION
# ==========================================
# Cek struktur dataset
print("Dataset Structure:")
print(os.listdir('/content/dataset'))

# Hitung jumlah dataset per kelas
def count_from_folder(directory):
    counts = {}
    for class_name in os.listdir(directory):
        class_path = os.path.join(directory, class_name)
        if os.path.isdir(class_path):
            counts[class_name] = len(os.listdir(class_path))
    return counts

df = pd.DataFrame({
    "Train": count_from_folder('/content/dataset/training'),
    "Validation": count_from_folder('/content/dataset/validation'),
    "Test": count_from_folder('/content/dataset/evaluation')
}).fillna(0).astype(int)

df.loc["Total"] = df.sum()
print("\nDataset Distribution:")
print(df)

# Visualisasi distribusi data
totals = df.loc["Total"]
plt.figure(figsize=(10, 5))
plt.bar(totals.index, totals.values, color=['#2c4fa1', '#a4befa', '#dd8046'])
plt.title("Distribusi Jumlah Data per Split", fontsize=14, fontweight='bold')
plt.ylabel("Jumlah Data")
plt.xlabel("Dataset Split")
for i, v in enumerate(totals.values):
    plt.text(i, v + 100, str(v), ha='center', fontweight='bold')
plt.tight_layout()
plt.savefig('dataset_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

# Visualisasi per kelas (Training)
plt.figure(figsize=(12, 6))
plt.bar(df.index[:-1], df["Train"][:-1], color='#2c4fa1')
plt.xticks(rotation=45, ha='right')
plt.title("Distribusi Data Training per Kelas", fontsize=14, fontweight='bold')
plt.ylabel("Jumlah Data")
plt.xlabel("Kelas")
plt.tight_layout()
plt.savefig('train_class_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

# Tampilkan contoh gambar
BASE_DIR = '/content/dataset/training'
classes = os.listdir(BASE_DIR)

plt.figure(figsize=(15, 5))
for i in range(5):
    cls = random.choice(classes)
    cls_path = os.path.join(BASE_DIR, cls)
    img_name = random.choice(os.listdir(cls_path))
    img_path = os.path.join(cls_path, img_name)
    img = Image.open(img_path)

    plt.subplot(1, 5, i + 1)
    plt.imshow(img)
    plt.title(cls, fontsize=12)
    plt.axis('off')

plt.suptitle('Contoh Gambar Dataset Food-11', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('sample_images.png', dpi=300, bbox_inches='tight')
plt.show()

# ==========================================
# 4. KONFIGURASI & PATH DATASET
# ==========================================
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 25

train_dir = '/content/dataset/training'
val_dir = '/content/dataset/validation'
eval_dir = '/content/dataset/evaluation'

# ==========================================
# 5. FUNGSI HELPER UNTUK EVALUASI
# ==========================================
def plot_training_history(history, model_name):
    """Plot training history: Loss dan Accuracy"""
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # Plot Loss
    axes[0].plot(history.history['loss'], label='Training Loss', linewidth=2)
    axes[0].plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
    axes[0].set_title(f'{model_name} - Loss', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    # Plot Accuracy
    axes[1].plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
    axes[1].plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
    axes[1].set_title(f'{model_name} - Accuracy', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Accuracy')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(f'{model_name}_training_history.png', dpi=300, bbox_inches='tight')
    plt.show()

def evaluate_model(model, test_generator, model_name):
    """Evaluasi model: Classification Report & Confusion Matrix"""
    print(f"\n{'='*60}")
    print(f"EVALUASI MODEL: {model_name}")
    print(f"{'='*60}\n")

    # Prediksi
    test_generator.reset()
    predictions = model.predict(test_generator, verbose=1)
    y_pred = np.argmax(predictions, axis=1)
    y_true = test_generator.classes

    # Classification Report
    class_names = list(test_generator.class_indices.keys())
    report = classification_report(y_true, y_pred, target_names=class_names, digits=4)
    print("Classification Report:")
    print(report)

    # Simpan classification report ke file
    with open(f'{model_name}_classification_report.txt', 'w') as f:
        f.write(f"Classification Report - {model_name}\n")
        f.write("="*60 + "\n\n")
        f.write(report)

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'label': 'Count'})
    plt.title(f'Confusion Matrix - {model_name}', fontsize=14, fontweight='bold')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.savefig(f'{model_name}_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Accuracy Score
    accuracy = accuracy_score(y_true, y_pred)
    print(f"\nTest Accuracy: {accuracy*100:.2f}%\n")

    return {
        'predictions': predictions,
        'y_pred': y_pred,
        'y_true': y_true,
        'accuracy': accuracy,
        'report': report
    }

# ==========================================
# 6. MODEL 1: CNN BASE (NON-PRETRAINED)
# ==========================================
print("\n" + "="*60)
print("MODEL 1: CNN BASE (NON-PRETRAINED)")
print("="*60 + "\n")

# Data Generator untuk CNN Base (tanpa preprocessing khusus)
train_datagen_cnn = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)

valid_datagen_cnn = ImageDataGenerator(rescale=1./255)
test_datagen_cnn = ImageDataGenerator(rescale=1./255)

train_dataset_cnn = train_datagen_cnn.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

valid_dataset_cnn = valid_datagen_cnn.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

test_dataset_cnn = test_datagen_cnn.flow_from_directory(
    eval_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# Arsitektur CNN Base
model_cnn = Sequential([
    # Block 1
    Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(224, 224, 3)),
    BatchNormalization(),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.25),

    # Block 2
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.25),

    # Block 3
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.25),

    # Block 4
    Conv2D(256, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    Conv2D(256, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.25),

    # Fully Connected Layers
    Flatten(),
    Dense(512, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(train_dataset_cnn.num_classes, activation='softmax')
], name='CNN_Base')

model_cnn.summary()

# Compile Model
model_cnn.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Callbacks
callbacks_cnn = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1),
    ModelCheckpoint('model_cnn_base.keras', save_best_only=True, monitor='val_accuracy', verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7, verbose=1)
]

# Training
history_cnn = model_cnn.fit(
    train_dataset_cnn,
    validation_data=valid_dataset_cnn,
    epochs=EPOCHS,
    callbacks=callbacks_cnn,
    verbose=1
)

# Plot Training History
plot_training_history(history_cnn, 'CNN_Base')

# Evaluasi Model
results_cnn = evaluate_model(model_cnn, test_dataset_cnn, 'CNN_Base')

# Simpan model final
model_cnn.save('model_cnn_base_final.keras')
print(" Model CNN Base disimpan: model_cnn_base_final.keras")

# ==========================================
# 7. MODEL 2: EfficientNetB7 (PRETRAINED)
# ==========================================
print("\n" + "="*60)
print("MODEL 2: EfficientNetB7 (PRETRAINED)")
print("="*60 + "\n")

# Data Generator untuk EfficientNetB7
train_datagen_b7 = ImageDataGenerator(
    preprocessing_function=preprocess_b7,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)

valid_datagen_b7 = ImageDataGenerator(preprocessing_function=preprocess_b7)
test_datagen_b7 = ImageDataGenerator(preprocessing_function=preprocess_b7)

train_dataset_b7 = train_datagen_b7.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

valid_dataset_b7 = valid_datagen_b7.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

test_dataset_b7 = test_datagen_b7.flow_from_directory(
    eval_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# Base Model: EfficientNetB7
base_b7 = EfficientNetB7(
    include_top=False,
    weights='imagenet',
    input_shape=(224, 224, 3)
)

# Fine-tuning: Unfreeze last 30 layers
base_b7.trainable = True
for layer in base_b7.layers[:-30]:
    layer.trainable = False

# Arsitektur Model
model_efficientnet = Sequential([
    base_b7,
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2),
    Conv2D(256, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    GlobalAveragePooling2D(),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(train_dataset_b7.num_classes, activation='softmax')
], name='EfficientNetB7')

model_efficientnet.summary()

# Compile Model
model_efficientnet.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Callbacks
callbacks_b7 = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1),
    ModelCheckpoint('model_efficientnet_b7.keras', save_best_only=True, monitor='val_accuracy', verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-7, verbose=1)
]

# Training
history_efficientnet = model_efficientnet.fit(
    train_dataset_b7,
    validation_data=valid_dataset_b7,
    epochs=EPOCHS,
    callbacks=callbacks_b7,
    verbose=1
)

# Plot Training History
plot_training_history(history_efficientnet, 'EfficientNetB7')

# Evaluasi Model
results_efficientnet = evaluate_model(model_efficientnet, test_dataset_b7, 'EfficientNetB7')

# Simpan model final
model_efficientnet.save('model_efficientnet_b7_final.keras')
print(" Model EfficientNetB7 disimpan: model_efficientnet_b7_final.keras")

# ==========================================
# 8. MODEL 3: MobileNetV2 (PRETRAINED)
# ==========================================
print("\n" + "="*60)
print("MODEL 3: MobileNetV2 (PRETRAINED)")
print("="*60 + "\n")

# Data Generator untuk MobileNetV2
train_datagen_mnet = ImageDataGenerator(
    preprocessing_function=preprocess_mnet,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)

valid_datagen_mnet = ImageDataGenerator(preprocessing_function=preprocess_mnet)
test_datagen_mnet = ImageDataGenerator(preprocessing_function=preprocess_mnet)

train_dataset_mnet = train_datagen_mnet.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

valid_dataset_mnet = valid_datagen_mnet.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

test_dataset_mnet = test_datagen_mnet.flow_from_directory(
    eval_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# Base Model: MobileNetV2
base_mnet = MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet'
)

# Freeze base model
base_mnet.trainable = False

# Arsitektur Model
model_mobilenet = Sequential([
    base_mnet,
    GlobalAveragePooling2D(),
    Dropout(0.5),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dense(train_dataset_mnet.num_classes, activation='softmax')
], name='MobileNetV2')

model_mobilenet.summary()

# Compile Model
model_mobilenet.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Callbacks
callbacks_mnet = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1),
    ModelCheckpoint('model_mobilenet_v2.keras', save_best_only=True, monitor='val_accuracy', verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7, verbose=1)
]

# Training
history_mobilenet = model_mobilenet.fit(
    train_dataset_mnet,
    validation_data=valid_dataset_mnet,
    epochs=EPOCHS,
    callbacks=callbacks_mnet,
    verbose=1
)

# Plot Training History
plot_training_history(history_mobilenet, 'MobileNetV2')

# Evaluasi Model
results_mobilenet = evaluate_model(model_mobilenet, test_dataset_mnet, 'MobileNetV2')

# Simpan model final
model_mobilenet.save('model_mobilenet_v2_final.keras')
print(" Model MobileNetV2 disimpan: model_mobilenet_v2_final.keras")

# ==========================================
# 9. PERBANDINGAN KETIGA MODEL
# ==========================================
print("\n" + "="*60)
print("TABEL PERBANDINGAN MODEL")
print("="*60 + "\n")

# Buat tabel perbandingan
comparison_df = pd.DataFrame({
    'Nama Model': ['CNN Base (Non-Pretrained)', 'EfficientNetB7 (Pretrained)', 'MobileNetV2 (Pretrained)'],
    'Test Accuracy': [
        f"{results_cnn['accuracy']*100:.2f}%",
        f"{results_efficientnet['accuracy']*100:.2f}%",
        f"{results_mobilenet['accuracy']*100:.2f}%"
    ],
    'Analisis': [
        'Model CNN dasar yang dibangun dari awal, memerlukan training lebih lama',
        'Model pretrained dengan arsitektur kompleks, performa tinggi dengan fine-tuning',
        'Model pretrained yang ringan dan cepat, cocok untuk deployment'
    ]
})

print(comparison_df.to_string(index=False))
print("\n")

# Simpan tabel ke CSV
comparison_df.to_csv('model_comparison.csv', index=False)
print(" Tabel perbandingan disimpan: model_comparison.csv\n")

# Visualisasi perbandingan akurasi
plt.figure(figsize=(10, 6))
accuracies = [results_cnn['accuracy']*100,
              results_efficientnet['accuracy']*100,
              results_mobilenet['accuracy']*100]
models = ['CNN Base', 'EfficientNetB7', 'MobileNetV2']
colors = ['#2c4fa1', '#dd8046', '#a4befa']

bars = plt.bar(models, accuracies, color=colors, edgecolor='black', linewidth=1.5)
plt.title('Perbandingan Akurasi Test Ketiga Model', fontsize=14, fontweight='bold')
plt.ylabel('Accuracy (%)')
plt.ylim(0, 100)

# Tambahkan nilai di atas bar
for bar, acc in zip(bars, accuracies):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{acc:.2f}%', ha='center', va='bottom', fontweight='bold', fontsize=12)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('model_comparison_chart.png', dpi=300, bbox_inches='tight')
plt.show()

# ==========================================
# 10. DOWNLOAD SEMUA FILE
# ==========================================
print("\n" + "="*60)
print("DOWNLOAD FILE UNTUK STREAMLIT")
print("="*60 + "\n")

# List file yang akan didownload
files_to_download = [
    'model_cnn_base_final.keras',
    'model_efficientnet_b7_final.keras',
    'model_mobilenet_v2_final.keras',
    'model_comparison.csv',
    'CNN_Base_classification_report.txt',
    'EfficientNetB7_classification_report.txt',
    'MobileNetV2_classification_report.txt',
    'CNN_Base_training_history.png',
    'EfficientNetB7_training_history.png',
    'MobileNetV2_training_history.png',
    'CNN_Base_confusion_matrix.png',
    'EfficientNetB7_confusion_matrix.png',
    'MobileNetV2_confusion_matrix.png',
    'model_comparison_chart.png',
    'dataset_distribution.png'
]

print("File yang tersedia untuk didownload:\n")
for i, file in enumerate(files_to_download, 1):
    if os.path.exists(file):
        print(f" {i}. {file}")
    else:
        print(f" {i}. {file} (tidak ditemukan)")

print("\n" + "="*60)
print("DOWNLOAD OTOMATIS DIMULAI...")
print("="*60 + "\n")

# Download semua file yang ada
for file in files_to_download:
    if os.path.exists(file):
        try:
            files.download(file)
            print(f" Downloaded: {file}")
        except Exception as e:
            print(f" Error downloading {file}: {str(e)}")

print("\n SEMUA FILE SIAP UNTUK STREAMLIT DASHBOARD!")
print("="*60)
```
